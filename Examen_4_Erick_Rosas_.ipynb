# **Pregunta 2**

## <small> a. Crea un DataFrame df_filtrado con todas las filas de df_diabetes donde Glucose > 120  </small>


```python
df_filtrado = df_diabetes[df_diabetes['Glucose'] > 120]
```

## <small> b. Agrega a df_filtrado una columna BMI_Category que clasifique el BMI  </small>

```python
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Bajo'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Sobrepeso'
    else:
        return 'Obeso'
```

```python
# Agregar columna BMI_Category
df_filtrado['BMI_Category'] = df_filtrado['BMI'].apply(categorize_bmi)
```

# **Pregunta 4**

## <small> a. Agrupa df_combinado por AgeGroup.  </small>

```python
df_agrupado = df_combinado.groupby('AgeGroup').size().reset_index(name='Count')
```

## <small> b. Calcula el promedio de Glucose y el promedio de BMI para cada grupo de edad.  </small>

```python
df_promedios = df_combinado.groupby('AgeGroup')[['Glucose', 'BMI']].mean().reset_index()
```

## <small> c. Ordena el resultado por el promedio de Glucose de mayor a menor  </small>

```python
df_promedios = df_promedios.sort_values(by='Glucose', ascending=False)
```
